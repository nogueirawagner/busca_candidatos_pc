```python
import pandas as pd

# Caminho do arquivo Excel
caminho_arquivo = 'C:\\Users\\engno\Desktop\\ClassificacaoAgentes\\ClassificacaoAgentes.xlsx'

# Lê o arquivo Excel
df = pd.read_excel(caminho_arquivo)

# Exibe as primeiras linhas do DataFrame
df.head()
```

```python
# Renomeando colunas específicas
agentes = df.rename(columns={'T#': 'Turma'})
# Converte a coluna de float para int
```

```python
agentes_pcd = agentes[
    (agentes['STATUS'] == 'AGUARDANDO') &
    (agentes['PCD'].notna())
]

agentes_pcd = agentes_pcd.sort_values(by='CG', ascending=True)
agentes_pcd
```

```python
ag_t1 = agentes[agentes["Turma"] == "T1"]
# Filtra os agentes que têm status diferente de 'convocado' e 'fim de fila'
# e que a coluna 'AC' não seja NaN

df_t1_nao_convocados = ag_t1[
    (ag_t1['STATUS'] == 'AGUARDANDO') &
    (ag_t1['AC'].notna())
]

df_t1_nao_convocados.reset_index(drop=True, inplace=True)

df_t1_nao_convocados = df_t1_nao_convocados.sort_values(by='CG', ascending=True)

# df_t1_nao_convocados = df_t1_nao_convocados[["ordem", "CG", "SBJDC", "Turma", "NOME"]]

# Exibe as primeiras linhas do DataFrame filtrado
df_t1_nao_convocados.head()
```

```python
# Filtra os agentes que têm status diferente de 'convocado' e 'fim de fila'
# e que a coluna 'AC' não seja NaN
t1_nao_convocados_negros = ag_t1[
    (ag_t1['STATUS'] == 'AGUARDANDO') &
    (ag_t1['PPP'].notna())
]

t1_nao_convocados_negros.reset_index(drop=True, inplace=True)

t1_nao_convocados_negros = t1_nao_convocados_negros.sort_values(by='CG', ascending=True)

# t1_nao_convocados_negros = t1_nao_convocados_negros[["ordem", "CG", "SBJDC", "Turma", "NOME"]]

# Exibe as primeiras linhas do DataFrame filtrado
t1_nao_convocados_negros.head()
```

```python
agentes_t2 = agentes[
    (agentes["Turma"] == "T2") &
    (agentes["STATUS"] == "AGUARDANDO") &
    (agentes['AC'].notna())
]
                     

agentes_t2.reset_index(drop=True, inplace=True)

agentes_t2 = agentes_t2.sort_values(by='CG', ascending=True)
agentes_t2.head()

```

```python
t2_sj = agentes_t2[
    (agentes_t2['STATUS'] == 'AGUARDANDO') &
    (agentes_t2['SBJDC'].notnull()) &
    (agentes_t2['AC'].notna())
]

t2_sj.head(15)
```

```python
agentes_t2_negros = agentes[
    (agentes["Turma"] == "T2") &
    (agentes["STATUS"] == "AGUARDANDO") &
    (agentes['PPP'].notna())
]
                     

agentes_t2_negros = agentes_t2_negros.sort_values(by='CG', ascending=True)
agentes_t2_negros.head()
```

```python
import numpy as np 
```

```python
import pandas as pd
from datetime import datetime, timedelta

# Data inicial
data_inicial = datetime(2024, 11, 6)

# Defina os feriados (excluindo o dia 15 de novembro)
feriados = [datetime(2024, 11, 15)]

# Cria uma regra de dias úteis customizada, excluindo o feriado especificado
dias_uteis_custom = pd.offsets.CustomBusinessDay(holidays=feriados)

# Gere as próximas 10 datas úteis a partir de 'data_inicial'
dias_uteis = pd.date_range(start=data_inicial, periods=40, freq=dias_uteis_custom).date
df_datas = pd.DataFrame({
    "Ordem": range(1, len(dias_uteis)),
    "Data_Inicial": dias_uteis[:-1],
    "Data_Final": dias_uteis[1:]
})
df_datas
```

```python
t1 = df_t1_nao_convocados
t1_ppp = t1_nao_convocados_negros
t2 = agentes_t2
t2_ppp = agentes_t2_negros
pcd = agentes_pcd

nome = 'naves'  # Em minúsculas

todos_ampla_cc = pd.concat([t1, t2], ignore_index=True)
todos_ampla_cc.reset_index(drop=True, inplace=True)
todos_ampla_cc['ordem'] = todos_ampla_cc.index + 1
todos_ampla_cc["Ampla"] = "Sim"
todos_ampla_cc["Negros"] = "Não"
todos_ampla_cc["PCD"] = "Não"

todos_negros = pd.concat([t1_ppp, t2_ppp], ignore_index=True)
todos_negros.reset_index(drop=True, inplace=True)
todos_negros['ordem'] = todos_negros.index + 1
todos_negros["Negros"] = "Sim"
todos_negros["Ampla"] = "Não"
todos_negros["PCD"] = "Não"

pcd.reset_index(drop=True, inplace=True)
pcd['ordem'] = pcd.index + 1
pcd["Ampla"] = "Não"
pcd["Negros"] = "Não"
pcd["PCD"] = "Sim"

qtd_agentes = 40
qtd_agentes_sem_pcd = 43
qtd_agentes_ppp = 11
qtd_agentes_pcd = 3

busca_nome_ampla = todos_ampla_cc[todos_ampla_cc['NOME'].str.contains(nome, case=False, na=False)]
busca_nome_ampla["Dia (Xº)"] = np.where(
    np.ceil(busca_nome_ampla['ordem'] / qtd_agentes).astype(int) <= 4,  # Condição
    np.ceil(busca_nome_ampla['ordem'] / qtd_agentes).astype(int),  # Se verdadeiro, usa qtd_agentes
    np.ceil(busca_nome_ampla['ordem'] / qtd_agentes_sem_pcd).astype(int)  # Se falso, usa qtd_agentes_sem_pcd
)
busca_nome_ampla["Data Provável"] = busca_nome_ampla["Dia (Xº)"].apply(lambda x: df_datas.loc[df_datas["Ordem"] == x, "Data_Inicial"].values[0] if x in df_datas["Ordem"].values else None)

busca_nome_negros = todos_negros[todos_negros['NOME'].str.contains(nome, case=False, na=False)]
busca_nome_negros["Dia (Xº)"] = np.ceil(busca_nome_negros['ordem'] / qtd_agentes_ppp).astype(int)
busca_nome_negros["Data Provável"] = busca_nome_negros["Dia (Xº)"].apply(lambda x: df_datas.loc[df_datas["Ordem"] == x, "Data_Inicial"].values[0] if x in df_datas["Ordem"].values else None)


busca_nome_pcd = pcd[pcd['NOME'].str.contains(nome, case=False, na=False)]
busca_nome_pcd["Dia (Xº)"] = np.ceil(busca_nome_pcd['ordem'] / qtd_agentes_pcd).astype(int)
busca_nome_pcd["Data Provável"] = busca_nome_pcd["Dia (Xº)"].apply(lambda x: df_datas.loc[df_datas["Ordem"] == x, "Data_Inicial"].values[0] if x in df_datas["Ordem"].values else None)

resultado_ampla = busca_nome_ampla[["NOME", "Dia (Xº)", "Data Provável", "ordem", "Ampla", "Negros", "PCD", "CG", "Turma"]]
resultado_negro = busca_nome_negros[["NOME", "Dia (Xº)", "Data Provável",  "ordem", "Ampla", "Negros", "PCD", "CG", "Turma"]]
resultado_pcd = busca_nome_pcd[["NOME", "Dia (Xº)", "Data Provável", "ordem", "Ampla", "Negros", "CG", "PCD", "Turma"]]

resultado_combinado = pd.concat([resultado_ampla, resultado_negro, resultado_pcd], ignore_index=True)

resultado_combinado
```
